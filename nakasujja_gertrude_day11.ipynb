```python
from sklearn.feature_selection import SelectKBest, chi2, f_classif, SelectFromModel, RFE
from sklearn.linear_model import LogisticRegression, Lasso
from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

df = pd.DataFrame(ng)

# Separate features (X) and target variable (y)
X = ng.drop('age', axis=1)
y = ng['age']

# Normalize Age to range 20-28
age_scaler = MinMaxScaler(feature_range=(20, 28))
ng['age'] = age_scaler.fit_transform(df[['age']].astype(float))

print("Normalized Age range:")
print(df[['age', 'age']].head(10))
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize using StandardScaler
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

X_standardized = pd.DataFrame(X_standardized, columns=X.columns)

print("Standardized features:")
print(X_standardized.head())
```

```python
# Example: Create polynomial features for 'age'
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(ng[['age']])

# Function to get polynomial feature names
def get_poly_feature_names(input_features, poly):
    feature_names = poly.get_feature_names(input_features)
    return [name.replace(" ", "*") for name in feature_names]

# Get feature names manually
poly_feature_names = get_poly_feature_names(ng[['age']].columns, poly)
X_poly = pd.DataFrame(X_poly, columns=poly_feature_names)

print("Polynomial features:")
print(X_poly.head())
```

```python
# Example: Create interaction features
interaction = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_interaction = interaction.fit_transform(X)
interaction_feature_names = interaction.get_feature_names(X.columns)
X_interaction = pd.DataFrame(X_interaction, columns=interaction_feature_names)

print("Interaction features:")
print(X_interaction.head())
```

```python
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# Define KFold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize lists to store evaluation metrics
fold_scores = []

# Loop over each fold
for fold, (train_index, val_index) in enumerate(kf.split(X, y)):
    X_train, X_val = X.iloc[train_index], X.iloc[val_index]
    y_train, y_val = y.iloc[train_index], y.iloc[val_index]
    
    # Initialize Decision Tree Regressor
    model = DecisionTreeRegressor(random_state=42)
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on validation set
    y_pred = model.predict(X_val)
    
    # Evaluate model (using Mean Squared Error as an example)
    fold_score = mean_squared_error(y_val, y_pred)
    fold_scores.append(fold_score)
    
    print(f"Fold {fold+1}: Mean Squared Error = {fold_score:.4f}")

# Optionally, aggregate results across folds
mean_score = np.mean(fold_scores)
print(f"\nAverage Mean Squared Error across {kf.n_splits} folds: {mean_score:.4f}")
```

```python
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Define KFold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize lists to store evaluation metrics
fold_scores = []

# Loop over each fold
for fold, (train_index, val_index) in enumerate(kf.split(X, y)):
    X_train, X_val = X.iloc[train_index], X.iloc[val_index]
    y_train, y_val = y.iloc[train_index], y.iloc[val_index]
    
    # Initialize Random Forest Regressor
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on validation set
    y_pred = model.predict(X_val)
    
    # Evaluate model (using Mean Squared Error as an example)
    fold_score = mean_squared_error(y_val, y_pred)
    fold_scores.append(fold_score)
    
    print(f"Fold {fold+1}: Mean Squared Error = {fold_score:.4f}")

# Optionally, aggregate results across folds
mean_score = np.mean(fold_scores)
print(f"\nAverage Mean Squared Error across {kf.n_splits} folds: {mean_score:.4f}")
```
