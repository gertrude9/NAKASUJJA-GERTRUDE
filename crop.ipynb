```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("crop.csv")

# Ensure the 'temperature' column is numeric
df['temperature'] = pd.to_numeric(df['temperature'], errors='coerce')

# Plot the histogram for temperature
plt.figure(figsize=(10, 6))
sns.histplot(df['temperature'], bins=30, kde=True)
plt.title('Distribution of Temperature')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv("crop.csv")

# List all column names
print(df.columns)

# Replace 'ActualTargetColumnName' with the actual column name of your target variable
target_column = 'rainfall'  # E.g., 'CropType'

# Check the distribution of the target variable
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x=target_column)
plt.title('Distribution of Target Variable')
plt.xlabel('rainfall')
plt.ylabel('Frequency')
plt.show()

# Print the value counts of the target variable
print(df[target_column].value_counts())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv("crop.csv")

# Exclude non-numeric columns (assuming 'CropType' is non-numeric)
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df_numeric = df[numeric_columns]

# Compute the correlation matrix
correlation_matrix = df_numeric.corr()

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df contains your dataset and 'N', 'K', 'P', 'temperature', 'humidity', 'ph', 'rainfall' are columns
Columns = ['N', 'K', 'P', 'temperature', 'humidity', 'ph', 'rainfall']

# Plotting boxplots for each column
plt.figure(figsize=(12, 8))
for i, column in enumerate(Columns, 1):
    plt.subplot(2, 4, i)
    sns.boxplot(data=df, x=column)
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)

plt.tight_layout()
plt.show()
```
