```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
df = pd.read_csv("niceguys.csv")

# Create a numerical representation for boyfriend/girlfriend (optional)
# If the column is already numerical (0/1), comment out this line
df["HasPartner"] = df["do_you_have_aboyfriend/girlfriend"].map({"No": 0, "Yes": 1})

# Filter data for those with partners
data_with_partner = df[df["HasPartner"] == 1]

# Analyze gender distribution for those with partners
genderCountsPartnered = data_with_partner["sex"].value_counts()


```

```python
# Step 1: Identify numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Step 2: Create a subset DataFrame with numeric columns only
df_numeric = df[numeric_columns]

# Step 3: Compute correlation matrix
correlation_matrix = df_numeric.corr()

# Step 4: Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Age and girlfriend or boyfriend
age_gf = df.groupby(['age', 'do_you_have_aboyfriend/girlfriend']).size().unstack().fillna(0)
print("\nAge and Girlfriend/Boyfriend distribution:")
print(age_gf)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  # Import matplotlib.pyplot

# Read the data from the CSV file (assuming you have this code earlier)
df = pd.read_csv("niceguys.csv")

# Boxplot of living allowance and presence of girlfriend/boyfriend
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='do_you_have_aboyfriend/girlfriend', y='living allowance')
plt.title('Living Allowance by Relationship Status')
plt.xlabel('Relationship Status (0 = No, 1 = Yes)')
plt.ylabel('Living Allowance')
plt.show()
```
