```python
import numpy as np # type: ignore
import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib as plt

df=pd.read_csv("niceguys.csv")
#display the first row of the dataframe
print (df.head(5))
```

```python

# method to return the number of rows that are empty
print("Missing values in each column:\n", df.isna().sum())
```

```python
# Check for missing values
print("\nMissing values after filling:")
print(df.isnull().sum())
```

```python
# Check for non-numeric values in 'living allowance' after cleaning
print("\nUnique values in 'living allowance' after cleaning:")
print(df['living allowance'].unique())
```

```python
# Convert the column to numeric (this will convert non-convertible values to NaN)
df['living allowance'] = pd.to_numeric(df['living allowance'], errors='coerce')
```

```python
# Fill missing values with the mean of the column
df['living allowance'].fillna(df['living allowance'].mean(), inplace=True)
```

```python
# Mode
mode_value = df["living allowance"].mode()
print(mode_value)

# df["living_allowance"].fillna(x,inplace = True)
df.fillna({"living allowance": mode_value}, inplace=True)
print(df)
```

```python

# Mean
mean_value = df["living allowance"].mean()
print(mean_value)

# df["living_allowance"].fillna(x,inplace = True)
df.fillna({"living allowance": mean_value}, inplace=True)
print(df)

```

```python
# Step 1: Identify numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Step 2: Create a subset DataFrame with numeric columns only
df_numeric = df[numeric_columns]

# Step 3: Compute correlation matrix
correlation_matrix = df_numeric.corr()

# Step 4: Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Install matplotlib


# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("cleaned_niceguys.csv")

# Group by age and count the occurrences
age_counts = df['age'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(10, 8))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(age_counts))))
plt.title('Age Distribution of Students')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
